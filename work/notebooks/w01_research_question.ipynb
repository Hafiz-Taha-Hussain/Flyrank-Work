## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Research question:** Given a content page's last-90-day search and engagement signals,
which pages should a content/SEO reviewer look at first this week, and what should they
consider doing (refresh, expand, protect, prune, or monitor)?

**Unit of analysis:** one content page (`content_id`), described by its trailing 90-day
metrics, pseudonymized to one `client_id`.

**Output:** a ranked queue of pages with a score, a suggested action, and a short reason
code the reviewer can inspect (e.g. "declining_with_demand" or "stale_visible_page").

**Who acts on it, and what do they do:** a content or SEO reviewer with limited weekly
capacity opens the top N pages on the queue and decides whether to refresh the content,
expand it, protect it as-is, prune it, or just keep monitoring it.

**Cost of a wrong call — and why the two error types aren't symmetric:**
- *False negative* (a genuinely declining, high-demand page never surfaces near the top of
  the queue): the page keeps losing visibility unaddressed, and nobody catches it until the
  damage is already done. This is the expensive direction.
- *False positive* (a stable or healthy page gets flagged and reviewed): costs a reviewer a
  few minutes checking a page that didn't need it. Annoying, but cheap and easily fixed with
  a better threshold or reason code.
- Because a missed decline is more costly than a wasted review, the project should lean
  toward not missing genuine opportunities, and precision@K (do the top K pages actually
  deserve review?) matters more than plain accuracy — it directly reflects the reviewer's
  real capacity and workflow.

**Why data or ML helps at all, rather than a simple if-statement:** a handful of individual
rules (stale + visible, declining + demand, thin + visible, page-one decay, low CTR, low
engagement) each catch *some* candidates on their own, but a page's real priority depends
on how several of these signals combine and trade off against each other — that's exactly
the kind of messy, many-signal, interacting pattern that's hard to hand-write as a single
rule but learnable from data. The starter pipeline's own baseline-vs-model comparison shows
this: the fixed rule alone (baseline_refresh_score) tops out at a modest precision@50, while
a model trained on the same signals does meaningfully better — evidence that there's a
real, learnable pattern in the combination of signals worth spending the next 7 weeks on,
not something a single if-statement already captures.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

**What this project will be able to say:**
- *Observed*: which measured signals (staleness, position, CTR, engagement, trend) are
  associated with pages later marked as declining, in this dataset, in this time window.
- *Directional*: pages with more of these risk signals stacked together tend to score
  higher in the review queue than pages with fewer.
- *Decision-support*: the ranked queue helps a reviewer spend limited time on the most
  promising candidates first — it prioritizes review, it does not replace it.

**What this project will never claim:**
- *Not causal*: a high score does not mean refreshing the page will cause it to recover.
  Proving that would require an actual experiment (e.g. A/B testing refreshed vs.
  untouched pages), which this observational dataset cannot provide.
- *Not a Google-algorithm claim*: I am not claiming to have discovered or predicted any
  search-engine ranking factor — only that certain observed signals correlate with
  observed outcomes in this data.
- *Not a guarantee*: "declining" here is a proxy label (`trend_direction == "down"`,
  calculated from the current window), not a confirmed future outcome — I'll treat it as
  a starting point to improve on, not as ground truth.
- *No private or identifying data*: no client names, URLs, domains, or raw queries will
  appear anywhere in this or future notebooks — only pseudonymized IDs and aggregated
  metrics.
